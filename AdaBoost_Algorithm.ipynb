```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df=df.drop('Cabin' , axis=1)
```

```python
from sklearn.impute import SimpleImputer
print(f'Number of Null values in Age Column before Imputation : {df.Age.isnull().sum()}')
si=SimpleImputer(strategy='mean')
df['Age']=si.fit_transform(df[['Age']])
print(f'Number of Null values in Age Column before Imputation : {df.Age.isnull().sum()}')
```

```python
df['Survived']=si.fit_transform(df[['Survived']])
```

```python
si1=SimpleImputer(strategy='most_frequent')
df['Embarked']=si1.fit_transform(df[['Embarked']]).ravel()
```

```python
df.isnull().sum()
```

```python
df['Fare']=si1.fit_transform(df[['Fare']]).ravel()
```

```python
df.isnull().sum()
```

```python
df=df.drop(['Name' , 'Ticket'] , axis=1)
```

```python
df
```

```python
threshold=3
from scipy.stats import zscore

numerical_features=['Fare' , 'Age']

z_score=np.abs(zscore(df[numerical_features]))
print(z_score)
```

```python
outliers=np.where(z_score >threshold)
print(outliers)
```

```python
from scipy.stats.mstats import winsorize

df['Age']=winsorize(df['Age'] , limits=[0.15 ,0.15])
df['Fare']=winsorize(df['Fare'] , limits=[0.15 ,0.15])
```

```python
from sklearn.preprocessing import OrdinalEncoder
oe=OrdinalEncoder()

df['Sex']=oe.fit_transform(df[['Sex']])
```

```python
df
```

```python
df['Embarked']=oe.fit_transform(df[['Embarked']])
```

```python
df['Survived']=oe.fit_transform(df[['Survived']])
```

```python
df
```

```python
X=df.drop('Survived' , axis=1)
y=df['Survived']
```

```python
from sklearn.model_selection import train_test_split
X_train , X_test , y_train, y_test=train_test_split(X,y , test_size=0.3, random_state=0)
```

```python
from sklearn.ensemble import AdaBoostClassifier

ab=AdaBoostClassifier(n_estimators=45 , learning_rate=1, random_state=0)

ab.fit(X_train, y_train)

y_pred=ab.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred , y_test)
```
